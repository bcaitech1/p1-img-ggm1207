```python
def change_2d_to_1d(tens):
    if len(tens.shape) == 2:
        tens = tens.reshape(-1)
    return tens

def change_age_to_cat(age_logit):
    # age_logit: 1d
    age_logit = torch.where(age_logit < 30, torch.zeros_like(age_logit), age_logit)
    age_logit = torch.where(age_logit >= 60, torch.ones_like(age_logit) * 2, age_logit)
    age_logit = torch.where(age_logit >= 30, torch.ones_like(age_logit), age_logit)
    
    return age_logit.type(torch.long)
```

## 정리

- 학습 할 때는 f1_score, accuracy 계산
- predict 할 때는 validation 에 대해서 계산 후 eval 각각 계산한 후 저장

```python
age_model = torch.load(age_path)
mask_model = torch.load(mask_path)
gender_model = torch.load(gender_path)
```

```python
images, labels = next(iter(train_dataloader))
images, labels = images.cuda(), labels

age_label = labels[1].cuda()
mask_label = labels[2].cuda()
gender_label = labels[0].cuda()

# 여기까지는 하지
```

```python
# Age 
from sklearn.metrics import f1_score

age_label = change_age_to_cat(age_label)

age_preds = age_model(images)
age_preds = change_2d_to_1d(age_preds)
age_preds = change_age_to_cat(age_preds)

f1_sco = cal_metrics(age_preds, age_label, f1_score)
acc = accuracy(age_preds, age_label)
print(f1_sco, acc)
```

```python
# Mask

mask_preds = mask_model(images)
mask_preds = torch.argmax(mask_preds, dim=1)
mask_preds = change_2d_to_1d(mask_preds)
f1_sco = cal_metrics(mask_preds, mask_label, f1_score)
acc = accuracy(mask_preds, mask_label)
print(f1_sco, acc)
```

```python
# Gender

gen_preds = gender_model(images)
gen_preds = torch.argmax(gen_preds, dim=1)
gen_preds = change_2d_to_1d(gen_preds)

f1_sco = cal_metrics(gen_preds, gender_label, f1_score)
acc = accuracy(gen_preds, gender_label)
print(f1_sco, acc)
```
