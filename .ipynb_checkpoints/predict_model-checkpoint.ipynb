```python
class EvalDataset(Dataset):
    def __init__(self, transform):
        df = pd.read_csv('/opt/ml/input/data/eval/info.csv')
        self.images = df['ImageID']
        self.images = list(map(lambda x: os.path.join('/opt/ml/input/data/eval/images', x), self.images))
        self.transform = transform
        
    def __getitem__(self, idx):
        image = self.images[idx]
        image = Image.open(image)
        
        if self.transform:
            image = transform(image)
            
        return image
            
    def __len__(self):
        return len(self.images)
```

```python
args = get_args()
transform = get_transforms(args)
eval_dataset = EvalDataset(transform)
```

```python
eval_dataloader = DataLoader(eval_dataset, batch_size=64, shuffle=False)
_, valid_dataloader = get_dataloader(args)
```
