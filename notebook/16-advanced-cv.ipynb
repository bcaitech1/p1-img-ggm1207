# 이미지 추출 시도해보기

```python
import os
from glob import glob
from PIL import Image
import matplotlib.pyplot as plt

import cv2
import albumentations as A

os.chdir('/opt/ml/P-Stage/1-STAGE/')
```

```python
sample_path = '/opt/ml/input/data/eval/images/0001b62fb1057a1182db2c839de232dbef0d1e90.jpg'
sample_image = cv2.imread(sample_path)
sample_image = cv2.cvtColor(sample_image, cv2.COLOR_BGR2RGB)
sample_image.shape
```

```python
plt.imshow(sample_image)
plt.show()
```

```python
sobelx = cv2.Sobel(sample_image, cv2.CV_8U, 1, 0, ksize=3)
sobely = cv2.Sobel(sample_image, cv2.CV_8U, 0, 1, ksize=3)
laplacian = cv2.Laplacian(sample_image, cv2.CV_8U)
```

```python
plt.figure(figsize=(18, 8))
plt.subplot(2, 2, 1)
plt.imshow(sample_image)
plt.title('Original')
plt.axis("off")
plt.subplot(2, 2, 2)
plt.imshow(laplacian)
plt.title('Laplacian')
plt.axis("off")
plt.subplot(2, 2, 3)
plt.imshow(sobelx)
plt.title('Sobel X')
plt.axis("off")
plt.subplot(2, 2, 4)
plt.imshow(sobely)
plt.title('Sobel Y')
plt.axis("off")
plt.tight_layout()
plt.show()
```
